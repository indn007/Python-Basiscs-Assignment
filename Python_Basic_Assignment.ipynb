### **Q10.  Write code to demonstrate type casting with list elements.**

```python
my_list = ['10', '20', '30', '40']

# Convert string elements to integers
int_list = [int(x) for x in my_list]
print("integer elements:", int_list)

# Convert string elements to floats
float_list = [float(x) for x in my_list]
print("float elements:", float_list)

# Convert string elements to booleans (assuming '0' as False and other strings as True)
bool_list = [bool(int(x)) for x in my_list]
print("boolean elements:", bool_list)
```

### **Q14.  Write a program to reverse a string using a while loop.**

```python
o_string = input("Enter the String: ")

r_string = ""

index = len(o_string) - 1

while index >= 0:
  r_string += o_string[index]
  index -= 1

print("Reversed String: ", r_string)
```

### **Q15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.**

```python
number = int(input("Enter a non-negative integer: "))

factorial = 1
current = number

if number < 0:
    print("Factorial is not defined for negative numbers.")
elif number == 0:
    print("The factorial of 0 is 1.")
else:
    while current > 0:
        factorial *= current
        current -= 1          # Decrease the current number by 1

    print(f"The factorial of {number} is {factorial}.")
```
